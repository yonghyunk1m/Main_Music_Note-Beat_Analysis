```python
import numpy as np
import librosa, librosa.display
import matplotlib.pyplot as plt
import music21
from music21 import chord
from IPython.display import Image
from collections import Counter

# Music Note - Frequency Table
NoteHz = [[27.5,'A0'],[29.1,'A#0'],[30.9,'B0'],[32.7,'C1'],[34.6,'C#1'],[36.7,'D1'],
          [38.9,'D#1'],[41.2,'E1'],[43.6,'F1'],[46.2,'F#1'],[49,'G1'],[52,'G#1'],
          [55,'A1'],[58.3,'A#1'],[61.7,'B1'],[65.4,'C2'],[69.3,'C#2'],[73.4,'D2'],
          [77.8,'D#2'],[82.4,'E2'],[87.3,'F2'],[92.5,'F#2'],[98,'G2'],[103.8,'G#2'],
          [110,'A2'],[116.5,'A#2'],[123.5,'B2'],[130.8,'C3'],[138.6,'C#3'],[146.8,'D3'],
          [155.6,'D#3'],[164.8,'E3'],[174.6,'F3'],[185,'F#3'],[196,'G3'],[207.6,'G#3'],
          [220,'A3'],[233.1,'A#3'],[247,'B3'],[261.6,'C4'],[277.2,'C#4'],[293.7,'D4'],
          [311.1,'D#4'],[329.6,'E4'],[349.2,'F4'],[370,'F#4'],[392,'G4'],[415.3,'G#4'],
          [440,'A4'],[466.2,'A#4'],[493.9,'B4'],[523.2,'C5'],[554.4,'C#5'],[587.3,'D5'],
          [622.2,'D#5'],[659.2,'E5'],[698.4,'F5'],[740,'F#5'],[784,'G5'],[830.6,'G#5'],
          [880,'A5'],[932.3,'A#5'],[987.8,'B5'],[1046.5,'C6'],[1108.7,'C#6'],[1174.6,'D6'],
          [1244.5,'D#6'],[1318.5,'E6'],[1396.9,'F6'],[1480,'F#6'],[1568,'G6'],[1661.2,'G#6'],
          [1760,'A6'],[1864.6,'A#6'],[1975.5,'B6'],[2093,'C7'],[2217.5,'C#7'],[2349.3,'D7'],
          [2489,'D#7'],[2637,'E7'],[2793.8,'F7'],[2960,'F#7'],[3136,'G7'],[3322.4,'G#7'],
          [3520,'A7'],[3729.3,'A#7'],[3951,'B7'],[4186,'C8'],[4434.9,'C#8'],[4698.6,'D8'],
          [4978,'D#8'],[5274,'E8'],[5587.6,'F8'],[5919.9,'F#8'],[6271.9,'G8'],[6644.9,'G#8'],
          [7040,'A8'],[7458.6,'A#8'],[7902.1,'B8']]

### MAJOR CHORDS
# Pseudo-Probability of C Major Chord
def CM (dic):
    CM = ['C','E','G']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in CM:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of C#/Db Major Chord
def DbM (dic):
    DbM = ['C#','F','G#']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in DbM:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of D Major Chord
def DM (dic):
    DM = ['D','F#','A']
    score = 0
    count = 0
    for key, val in dic.items():
        if key in DM:
            score += val
            count += 1
    if count == 3:
        return score
    else:
        return 0

# Pseudo-Probability of D#/Eb Major Chord
def EbM (dic):
    EbM = ['D#','G4','A#']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in EbM:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of E Major Chord
def EM (dic):
    EM = ['E','G#','B']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in EM:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of F Major Chord
def FM (dic):
    FM = ['F','A','C']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in FM:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of F#/Gb Major Chord
def GbM (dic):
    GbM = ['F#','A#','C#']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in GbM:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of G Major Chord
def GM (dic):
    GM = ['G','B','D']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in GM:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of G#/Ab Major Chord
def AbM (dic):
    AbM = ['G#','C','D#']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in AbM:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of A Major Chord
def AM (dic):
    AM = ['A','C#','E']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in AM:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of A#/Bb Major Chord
def BbM (dic):
    BbM = ['A#','D','F']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in BbM:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of B Major Chord
def BM (dic):
    BM = ['B','D#','F#']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in BM:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0


### minor CHORDS
# Pseudo-Probability of C minor Chord
def Cm (dic):
    Cm = ['C','D#','G']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in Cm:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of C#/Db minor Chord
def Dbm (dic):
    Dbm = ['C#','E','G#']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in Dbm:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of D minor Chord
def Dm (dic):
    Dm = ['D','F','A']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in Dm:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of D#/Eb minor Chord
def Ebm (dic):
    Ebm = ['D#','F#','A#']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in Ebm:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of E minor Chord
def Em (dic):
    Em = ['E','G','B']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in Em:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of F minor Chord
def Fm (dic):
    Fm = ['F','G#','C']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in Fm:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of F#/Gb minor Chord
def Gbm (dic):
    Gbm = ['F#','A','C#']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in Gbm:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of G minor Chord
def Gm (dic):
    Gm = ['G','A#','D']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in Gm:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of G#/Ab minor Chord
def Abm (dic):
    Abm = ['G#','B','D#']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in Abm:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of A minor Chord
def Am (dic):
    Am = ['A','C','E']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in Am:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0


# Pseudo-Probability of A#/Bb minor Chord
def Bbm (dic):
    Bbm = ['A#','A#','F']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in Bbm:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

# Pseudo-Probability of B minor Chord
def Bm (dic):
    Bm = ['B','D','F#']
    score = 0
    count = 0
    for key, val in dic.items():
         if key in Bm:
                score += val
                count += 1
    if count == 3:
        return(score)
    else:
        return 0

    
##### COMMON CHORD PROGRESSION (Referred to https://www.guitar-chords.org.uk/chords-key-b.html)
### Major Key uses I, ii, IV, V, vi chords for their progression in common.

def CMajor_chords (dic):
    # Common Chords in C Major: C, Dm, F, G, Am
    Score = (CM(dic) + Dm(dic) + FM(dic) + GM(dic) + Am(dic))/5
    return Score
    
def DbMajor_chords (dic):
    # Common Chords in C# Major: C#, D#m, F#, G#, A#m
    Score = (DbM(dic) + Ebm(dic) + GbM(dic) + AbM(dic) + Bbm(dic))/5
    return Score

def DMajor_chords (dic):
    # Common Chords in D Major: D, Em, G, A, Bm
    Score = (DM(dic) + Em(dic) + GM(dic) + AM(dic) + Bm(dic))/5
    return Score

def EbMajor_chords (dic):
    # Common Chords in Eb Major: EbM, Fm, AbM, BbM, Cm
    Score = (EbM(dic) + Fm(dic) + AbM(dic) + BbM(dic) + Cm(dic))/5
    return Score

def EMajor_chords (dic):
    # Common Chords in E Major: EM, F#m, AM, BM, C#m
    Score = (EM(dic) + Gbm(dic) + AM(dic) + BM(dic) + Dbm(dic))/5
    return Score

def FMajor_chords (dic):
    # Common Chords in F Major: FM, Gm, BbM, CM, Dm
    Score = (FM(dic) + Gm(dic) + BbM(dic) + CM(dic) + Dm(dic))/5
    return Score

def GbMajor_chords (dic):
    # Common Chords in F# Major: F#M, G#m, BM, C#M, D#m
    Score = (GbM(dic) + Abm(dic) + BM(dic) + DbM(dic) + Ebm(dic))/5
    return Score

def GMajor_chords (dic):
    # Common Chords in G Major: GM, Am, CM, DM, Em
    Score = (GM(dic) + Am(dic) + CM(dic) + DM(dic) + Em(dic))/5
    return Score

def AbMajor_chords (dic):
    # Common Chords in Ab Major: AbM, Bbm, DbM, EbM, Fm
    Score = (AbM(dic) + Bbm(dic) + DbM(dic) + EbM(dic) + Fm(dic))/5
    return Score

def AMajor_chords (dic):
    # Common Chords in A Major: AM, Bm, DM, EM, F#m
    Score = (AM(dic) + Bm(dic) + DM(dic) + EM(dic) + Gbm(dic))/5
    return Score

def BbMajor_chords (dic):
    # Common Chords in Bb Major: BbM, Cm, EbM, FM, Gm
    Score = (BbM(dic) + Cm(dic) + EbM(dic) + FM(dic) + Gm(dic))/5
    return Score

def BMajor_chords (dic):
    # Common Chords in B Major: B, C#m, EM, F#M, G#m
    Score = (BM(dic) + Dbm(dic) + EM(dic) + GbM(dic) + Abm(dic))/5
    return Score

### minor Key uses i, ii, III, iv, v, VI, VII chords for their progression in common.    

def Cminor_chords (dic):
    # Common Chords in C minor: Cm, Dm, EbM, Fm, Gm, AbM, BbM
    Score = (Cm(dic) + Dm(dic) + EbM(dic) + Fm(dic) + Gm(dic)+ AbM(dic)+ BbM(dic))/7
    return Score

def Dbminor_chords (dic):
    # Common Chords in C# minor: C#m, D#m, EM, F#m, G#m, AM, BM
    Score = (Dbm(dic) + Ebm(dic) + EM(dic) + Gbm(dic) + Abm(dic)+ AM(dic)+ BM(dic))/7
    return Score

def Dminor_chords (dic):
    # Common Chords in D minor: Dm, Em, FM, Gm, Am, BbM, CM
    Score = (Dm(dic) + Em(dic) + FM(dic) + Gm(dic) + Am(dic)+ BbM(dic)+ CM(dic))/7
    return Score

def Ebminor_chords (dic):
    # Common Chords in Eb minor: Ebm, Fm, GbM, Abm, Bbm, BM, DbM
    Score = (Ebm(dic) + Fm(dic) + GbM(dic) + Abm(dic) + Bbm(dic)+ BM(dic)+ DbM(dic))/7
    return Score

def Eminor_chords (dic):
    # Common Chords in E minor: Em, F#m, GM, Am, Bm, CM, DM
    Score = (Em(dic) + Gbm(dic) + GM(dic) + Am(dic) + Bm(dic)+ CM(dic)+ DM(dic))/7
    return Score

def Fminor_chords (dic):
    # Common Chords in F minor: Fm, Gm, AbM, Bbm, Cm, DbM, EbM
    Score = (Fm(dic) + Gm(dic) + AbM(dic) + Bbm(dic) + Cm(dic)+ DbM(dic)+ EbM(dic))/7
    return Score

def Gbminor_chords (dic):
    # Common Chords in F# minor: F#m, G#m, AM, Bm, C#m, DM, EM
    Score = (Gbm(dic) + Abm(dic) + AM(dic) + Bm(dic) + Dbm(dic)+ DM(dic)+ EM(dic))/7
    return Score

def Gminor_chords (dic):
    # Common Chords in G minor: Gm, Am, BbM, Cm, Dm, EbM, FM
    Score = (Gm(dic) + Am(dic) + BbM(dic) + Cm(dic) + Dm(dic)+ EbM(dic)+ FM(dic))/7
    return Score

def Abminor_chords (dic):
    # Common Chords in Ab minor: Abm, Bbm, BM, Dbm, Ebm, EM, GbM
    Score = (Abm(dic) + Bbm(dic) + BM(dic) + Dbm(dic) + Ebm(dic)+ EM(dic)+ GbM(dic))/7
    return Score

def Aminor_chords (dic):
    # Common Chords in A minor: Am, Bm, CM, Dm, Em, FM, GM
    Score = (Am(dic) + Bm(dic) + CM(dic) + Dm(dic) + Em(dic)+ FM(dic)+ GM(dic))/7
    return Score

def Bbminor_chords (dic):
    # Common Chords in Bb minor: Bbm, Cm, DbM, Ebm, Fm, GbM, AbM
    Score = (Bbm(dic) + Cm(dic) + DbM(dic) + Ebm(dic) + Fm(dic)+ GbM(dic)+ AbM(dic))/7
    return Score

def Bminor_chords (dic):
    # Common Chords in B minor: Bm, C#m, DM, Em, F#m, GM, AM
    Score = (Bm(dic) + Dbm(dic) + DM(dic) + Em(dic) + Gbm(dic)+ GM(dic)+ AM(dic))/7
    return Score

# TOTAL SCORE
def KeyClassifier (dic):
    # 50:50 Score Distribution to Progression and Chord itself
    temp = [["CM",CMajor_chords(dic)+CM(dic)],["DbM",DbMajor_chords(dic)+DbM(dic)],["DM",DMajor_chords(dic)+DM(dic)]
            ,["EbM",EbMajor_chords(dic)+EbM(dic)],["EM",EMajor_chords(dic)+EM(dic)],["FM",FMajor_chords(dic)+FM(dic)]
            ,["GbM",GbMajor_chords(dic)+GbM(dic)],["GM",GMajor_chords(dic)+GM(dic)],["AbM",AbMajor_chords(dic)+AbM(dic)]
            ,["AM",AMajor_chords(dic)+AM(dic)],["BbM",BbMajor_chords(dic)+BbM(dic)],["BM",BMajor_chords(dic)+BM(dic)]
            ,["Cm",Cminor_chords(dic)+Cm(dic)],["Dbm",Dbminor_chords(dic)+Dbm(dic)],["Dm",Dminor_chords(dic)+Dm(dic)]
            ,["Ebm",Ebminor_chords(dic)+Ebm(dic)],["Em",Eminor_chords(dic)+Em(dic)],["Fm",Fminor_chords(dic)+Fm(dic)]
            ,["Gbm",Gbminor_chords(dic)+Gbm(dic)],["Gm",Gminor_chords(dic)+Gm(dic)],["Abm",Abminor_chords(dic)+Abm(dic)]
            ,["Am",Aminor_chords(dic)+Am(dic)],["Bbm",Bbminor_chords(dic)+Bbm(dic)],["Bm",Bminor_chords(dic)+Bm(dic)]]
    ans = ""
    maximum = 0
    for i in temp:
            if i[1] > maximum:
                ans = i[0]
                maximum = i[1]

    return ans

# Major or Minor #얘두 역시 Progression을 따라 가는게 맞음. Chord 자체보다는 곡 전체 분위기니깐!! (큰수의 법칙... Sampling 작은 N필요?)
def MmClassifier (dic):
    Major = CMajor_chords(dic)+DbMajor_chords(dic)+DMajor_chords(dic)+EbMajor_chords(dic)+EMajor_chords(dic)+FMajor_chords(dic)+GbMajor_chords(dic)+GMajor_chords(dic)+AbMajor_chords(dic)+AMajor_chords(dic)+BbMajor_chords(dic)+BMajor_chords(dic)
    minor = Cminor_chords(dic)+Dbminor_chords(dic)+Dminor_chords(dic)+Ebminor_chords(dic)+Eminor_chords(dic)+Fminor_chords(dic)+Gbminor_chords(dic)+Gminor_chords(dic)+Abminor_chords(dic)+Aminor_chords(dic)+Bbminor_chords(dic)+Bminor_chords(dic)
    if Major+minor != 0:
        print("This Music has ",round(100*Major/(Major+minor)),"% of Major mood, and ",round(100*minor/(Major+minor)),"% of minor mood.")
    else:
        print("This Music has 50% of Major mood and 50% of minor mood.")

def Top_Notes_wo_octaves (li, N):
    filtered_freq_seq = li[:N]; # For fast computation, filter the top N frequent Notes from the music. 
    Top_Notes = []
    for i in filtered_freq_seq:
        for idx, val in enumerate(NoteHz):
            if val[0] == i:
                Top_Notes.append(val[1])
            elif val[0] > i:
                avg = (NoteHz[idx-1][0]+val[0])/2
                if avg > i:
                    Top_Notes.append(NoteHz[idx-1][1])
                else:
                    Top_Notes.append(val[1])
                break

    # 더 근접한 Frequency에 해당하는 음으로 재배정한 Top Notes들의 Dictionray 생성
    Top_Notes_dic = {}

    for idx, val in enumerate(Top_Notes):
        if val not in Top_Notes_dic:
            Top_Notes_dic[val] = (len(Top_Notes)*2-idx)/len(Top_Notes) # 1~2 가중치
        else:
            Top_Notes_dic[val] += (len(Top_Notes)*2-idx)/len(Top_Notes)

    # Disregard the Octave for checking the Chord Combination
    Top_Notes_wo_octaves_dic = {}
    for key, val in Top_Notes_dic.items():
        temp = key[:len(key)-1]
        if temp not in Top_Notes_wo_octaves_dic:
            Top_Notes_wo_octaves_dic[temp] = val
        else:
            Top_Notes_wo_octaves_dic[temp] += val
    return Top_Notes_wo_octaves_dic
    
def Music_Analyzer (string):
    print("ANALYSIS OF \"", string,"\"")
    file = string # Read WAV Music file
    y, sr = librosa.load(file) #y: Amplitude, sr: sampling rate (22050)

    plt.figure(figsize=(15,5))
    plt.subplot(1,2,1)
    librosa.display.waveplot(y, sr, alpha=0.5)
    plt.xlabel("Time (mm:ss)")
    plt.ylabel("Amplitude")
    plt.title("Waveform")

    Y = np.fft.fft(y)
    Y_mag = np.abs(Y) 
    freq = np.linspace(0,sr,len(Y_mag)) # Maximum range of freq: 22050/2=11025Hz

    left_spectrum = Y_mag[:int(len(Y_mag)/2)]
    left_f = freq[:int(len(Y_mag)/2)]

    plt.subplot(1,2,2)
    plt.plot(left_f, left_spectrum)
    plt.xlabel("Frequency")
    plt.ylabel("Magnitude")
    plt.title("Power spectrum")

    max_freq = np.argmax(left_spectrum)
    length = len(left_f)
    divisor = length/11025

    # Play Main Music Note (The most frequently detected note)
    n = music21.note.Note(type='whole')
    n.pitch.frequency = max_freq/divisor
    print("[Main Note]")
    n.show()
    n.show('midi')

    frequent_seq = np.flip(np.argsort(left_spectrum)//divisor) # Change to real frequency values and sorted by descending order
 
    Top_Notes_wo_octaves_dic = Top_Notes_wo_octaves (frequent_seq, 50000)
    Top_Notes = sorted(Top_Notes_wo_octaves_dic.items(), key = lambda item: item[1], reverse = True)

    Top_Notes_wo_octaves_str = ""
    for i in Top_Notes:
        Top_Notes_wo_octaves_str += i[0]
        Top_Notes_wo_octaves_str += "4 "

    Top_4_Notes_str = "" # Top 4 Notes
    space_count = 0
    for i in Top_Notes_wo_octaves_str:
        if i == " ":
            space_count +=1
        if space_count == 4:
            break
        else:
            Top_4_Notes_str += i

    print("[Estimated Main 4 Notes]")
    notes = chord.Chord(Top_4_Notes_str, type='whole') 
    notes.show()
    notes.show('midi')

    Candidate_1 = KeyClassifier(Top_Notes_wo_octaves (frequent_seq, 5000))
    Candidate_2 = KeyClassifier(Top_Notes_wo_octaves (frequent_seq, 10000))
    Candidate_3 = KeyClassifier(Top_Notes_wo_octaves (frequent_seq, 50000))

    keys = Counter([Candidate_1, Candidate_2, Candidate_3]).most_common()
    if len(keys) == 1:
        print("This Music has Following Main Key:", Candidate_1)
    elif len(keys) == 2:
        print("This Music has Following Main Key:", Candidate_1)
        print("Also, this key can be the candidate:", keys[0][0])
    else:
        for i in len(keys):
            print("This Music has Following Main Keys:", keys[i][0])

    # Compute local onset autocorrelation
    hop_length = 512
    oenv = librosa.onset.onset_strength(y=y, sr=sr, hop_length=hop_length)
    tempogram = librosa.feature.tempogram(onset_envelope=oenv, sr=sr,
                                          hop_length=hop_length)
    # Compute global onset autocorrelation
    ac_global = librosa.autocorrelate(oenv, max_size=tempogram.shape[0])
    ac_global = librosa.util.normalize(ac_global)
    # Estimate the global tempo for display purposes
    tempo = librosa.beat.tempo(onset_envelope=oenv, sr=sr, hop_length=hop_length)[0]
    fig, ax = plt.subplots(1,2, figsize=(15, 5))
    times = librosa.times_like(oenv, sr=sr, hop_length=hop_length)
    ax[0].plot(times, oenv, label='Onset strength')
    ax[0].label_outer()
    ax[0].legend(frameon=True)
    librosa.display.specshow(tempogram, sr=sr, hop_length=hop_length, x_axis='time', y_axis='tempo', cmap='magma', ax=ax[1])
    ax[1].axhline(tempo, color='w', linestyle='--', alpha=1,
                label='Estimated tempo={:g}'.format(tempo))
    ax[1].legend(loc='upper right')
    ax[1].set(title='Tempogram')

    tempo2,_=librosa.beat.beat_track(y,sr=sr) # Ellis, Daniel PW. "Beat tracking by dynamic programming."
    if tempo != tempo2:
        print("The tempo of music can be ",tempo," or ", tempo2,"BPM")
    else:
        print("This Music has Following Tempo: ", round(tempo, 4), "BPM")
```
